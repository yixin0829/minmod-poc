Tested using Unstructured for parsing PDFs. Overall its performance on parsing correct table structure is not up to standard with Adobe PDF extraction API. Otherwise it's all good. Can be considered a candidate in case we need to switch to open-source PDF parser.

```python
from unstructured.partition.pdf import partition_pdf

PDF_PATH = "../data/raw/reports_processed/Bongará Zn 3-2019.pdf"

elements = partition_pdf(filename=PDF_PATH,
                         extract_images_in_pdf=False,
                         infer_table_structure=True,
                         strategy="hi_res"
                         chunking_strategy="by_title",
                         max_characters=4000,
                         new_after_n_chars=3800,
                         combine_text_under_n_chars=2000)

elements_fast = partition_pdf(PDF_PATH, strategy="fast")
```

```python
from collections import Counter

display(Counter(type(element) for element in elements))
print("")
# The composition of elements can be different for elements derived with the "fast" strategy
display(Counter(type(element) for element in elements_fast))
```
